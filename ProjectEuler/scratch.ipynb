## Amicable numbers

```python
PDDict = dict()
def getProperDivisors(n):
    if n in PDDict:
        return PDDict[n]
    else:
        properDivisors = []
        for i in range(1,int(n/2 + 1)):
            if n % i == 0:
                properDivisors.append(i)
        PDDict[n] = properDivisors
        return properDivisors
```

```python
amicableNums = set()

for n in range(1,10000 + 1):
    SUM = sum(getProperDivisors(n))    
    if (sum(getProperDivisors(SUM)) == n) and n != SUM:
        amicableNums.add(n)
        amicableNums.add(SUM)
        
print(len(amicableNums))
print(sum(amicableNums))
    
amicableNums
```

## Distinct primes factors

```python
import math
PrimeFactors = dict()
# List Version
def getPrimeFactors(n):
    if n in PrimeFactors:
        return PrimeFactors[n]
    
    else:
        for i in range(2, int(n/2+1)):
            if n % i == 0:
                PFs = getPrimeFactors(i) + getPrimeFactors(int(n/i))
                PrimeFactors[n] = PFs
                return PrimeFactors[n]
        PrimeFactors[n] = [n]
        return PrimeFactors[n]
  

# Set Version
import math
PrimeFactors = dict()
def getPrimeFactors(n):
    if n in PrimeFactors:
        return PrimeFactors[n]
    
    else:
        for i in range(2, int(n/2+1)):
            if n % i == 0:
                PFs = set()
                PFs.update(getPrimeFactors(i))
                PFs.update(getPrimeFactors(int(n/i)))
                PrimeFactors[n] = PFs
                return PrimeFactors[n]
            
        PrimeFactors[n] = set()
        PrimeFactors[n].add(n)
        return PrimeFactors[n]
```

## Triangular, pentagonal, and hexagonal

```python
def generatePentagonNumbers(rng):
    PentNumSet = set()
    for n in range(1, rng):
        num = int(n * (3*n-1)/2)
        PentNumSet.add(num)
    return(PentNumSet)


def generateTriangleNumbers(rng):
    TriNumSet = set()
    for n in range(1, rng):
        num = int(n * (n+1)/2)
        TriNumSet.add(num)
    return(TriNumSet)


def generateHexagonNumbers(rng):
    HexNumSet = set()
    for n in range(1, rng):
        num = int(n * (2*n-1))
        HexNumSet.add(num)
    return(HexNumSet)
```

## Double-base palindromes

```python
def reverseNum(n):
    n = str(n)
    new = []
    for i in range(1, len(n) + 1):
        new.append(n[-i])
    
    num = str(new[0])
    for i in range(1, len(new)):
        num = num + str(new[i])
    
    return int(num)
    
def isPalindrome(n):
    return (n == reverseNum((n)))
```

```python
SUM = 0
for i in range(0, 1000000):
    if isPalindrome(i) and isPalindrome(int(bin(i)[2:])):
        SUM += i
SUM 
```

##  Digit factorials

```python
F = dict()

def factorial(n):
    if n in F:
        return F[n]
    if n == 0:
        F[n] = 1
        return 1
    else:
        F[n] = n * factorial(n-1)
        return F[n]
    

def factorialize(num):
    numString = str(num)
    SUM = 0
    for i in range(len(numString)):
        SUM += factorial(int(numString[i]))
    return SUM
        
    
```

```python
count = 0
for num in range(1, 1000000):
    if factorialize(num) == num:
        print(num)
        count += 1
        
count + 2
print("FINISHED")
```

## Coin sums

```python
coins = [200, 100, 50, 20, 10, 5, 2 ,1]
Cost = 200
```

```python
def combinations(cost, lst):
    
    if cost == 0:
        return 1
    elif len(lst) == 0:
        return 0
    else:
    
        iterations = int(cost / lst[0]) + 1
        num = 0
        for i in range(iterations):
            num += combinations(cost - lst[0]*i, lst[1:])
        return num
```

```python
from functools import reduce

def factors(n):    
    return set(reduce(list.__add__, 
                ([i, n//i] for i in range(1, int(pow(n, 0.5) + 1)) if n % i == 0)))

# Generating Triangle Numbers
TriNums = [0]

for i in range(1, 200000000):
    TriNums.append(TriNums[-1] + i)
    L = len(factors(TriNums[-1]))
    if L > 500:
        print(TriNums[-1], len(factors(TriNums[-1])))
    
    
    
```

## Largest palindrome product

```python
def reverseNum(n):
    n = str(n)
    new = []
    for i in range(1, len(n) + 1):
        new.append(n[-i])
    
    num = str(new[0])
    for i in range(1, len(new)):
        
        num = num + str(new[i])
    
    return int(num)
    
    
    
def isPalindrome(n):
    return (n == reverseNum((n)))


collection = []
for i in range(10):
    for j in range(10):
        for k in range(10):
            
            for x in range(10):
                for y in range(10):
                    for z in range(10):
                        
                        num1 = 100*i + 10*j + k
                        num2 = 100*x + 10*y + z
                        
                        prod = num1 * num2
                        
                        if isPalindrome(prod):
                            collection.append(prod)
max(collection)
                        
                        
            
```

## Largest prime factor

```python
def largest_prime_factor(n):
    i = 2
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
    return n
```

## Non-abundant sums

```python
# Brute Force finding all the proper divisors of numbers
upper = 28123

divisors = []
for i in range(upper):
    divisors.append(set())

for num in range(1, upper):
    for j in range(1, num):
        if num % j == 0:
            divisors[num].add(j)
            
# Getting the Abundant Numbers
abundantNum = []
for i in range(len(divisors)):
    if sum(list(divisors[i])) > i:
        abundantNum.append(i)
len(abundantNum)
    
    
# Summing Abundant Nums
answer_nums = set()
for i in range(len(abundantNum)):
    print(i)
    for j in range(i, len(abundantNum)):
        if abundantNum[i] + abundantNum[j] < upper:
            answer_nums.add(abundantNum[i] + abundantNum[j])

ans = 0
for i in range(upper):
    if i not in answer_nums:
        ans += i
ans
    
```
